```python
import pandas as pd
import json
import os
from sqlalchemy import create_engine # DataBase Connection
```

```python
class PhonePe:

  def __init__(self,filePath=""): # Constructor Method
    self.filepath=filePath
    print(f"Hi {self.filepath}, Welcome To PhonePe")


  def Load_Transaction_SourceData(self,path): # Instance Method
    #path="D:/Guvi_Kannan/Data/data/aggregated/transaction/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list
    clm={'State':[], 'Year':[],'Quarter':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
                for z in D['data']['transactionData']:
                  Name=z['name']
                  count=z['paymentInstruments'][0]['count']
                  amount=z['paymentInstruments'][0]['amount']
                  clm['Transacion_type'].append(Name)
                  clm['Transacion_count'].append(count)
                  clm['Transacion_amount'].append(amount)
                  clm['State'].append(i)
                  clm['Year'].append(j)
                  clm['Quarter'].append(int(k.strip('.json')))
    #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(clm)
    print("Successfully retrieved data")
    BulkloadDatatoTable("aggtransaction",Agg_Trans)

  def Load_Users_SourceData(self,path): # Instance Method
    #path="D:/Guvi_Kannan/Data/data/aggregated/user/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list
    #Agg_state_list--> to get the list of states in India

    #<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

    #This is to extract the data's to create a dataframe

    users={'State':[], 'Year':[],'Quarter':[],'Brand':[], 'RegisteredCount':[], 'RegisteredPercentage':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
              # print(D['data']['usersByDevice'])
                if D['data']['usersByDevice'] is not None:
                    for z in D['data']['usersByDevice']:
                        brand=z['brand']
                        registeredCount=z['count']
                        registeredPercentage=z['percentage']
                        users['Brand'].append(brand)
                        users['RegisteredCount'].append(registeredCount)
                        users['RegisteredPercentage'].append(registeredPercentage)
                        users['State'].append(i)
                        users['Year'].append(j)
                        users['Quarter'].append(int(k.strip('.json')))
                  # print(brand)
      #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(users)
    BulkloadDatatoTable("aggusersbybrand",Agg_Trans)
  
  def Load_Insurance_SourceData(self,path): # Instance Method
    #path="D:/Guvi_Kannan/Data/data/aggregated/user/country/india/state/"
    Agg_state_list=os.listdir(path)
    Agg_state_list
    #Agg_state_list--> to get the list of states in India

    #<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

    #This is to extract the data's to create a dataframe

    Insurance={'State':[], 'Year':[],'Quarter':[],'InsuranceType':[], 'Insurance_count':[], 'Insurance_amount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
              # print(D['data']['usersByDevice'])
                if D['data']['transactionData'] is not None:
                    for z in D['data']['transactionData']:
                        type=z['paymentInstruments'][0]['type']
                        count=z['paymentInstruments'][0]['count']
                        amount=z['paymentInstruments'][0]['amount']
                        Insurance['InsuranceType'].append(type)
                        Insurance['Insurance_count'].append(count)
                        Insurance['Insurance_amount'].append(amount)
                        Insurance['State'].append(i)
                        Insurance['Year'].append(j)
                        Insurance['Quarter'].append(int(k.strip('.json')))
                  # print(brand)
      #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(Insurance)
    Agg_Trans
    BulkloadDatatoTable("AggInsurance",Agg_Trans)

  def Load_Top_TransactionInsurance(self,path,tableName): # Instance Method
    Agg_state_list=os.listdir(path)
    Agg_state_list
    #Agg_state_list--> to get the list of states in India

    #<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

    #This is to extract the data's to create a dataframe

    Top_transac={'State':[], 'Year':[],'Quarter':[],'District':[], 'Count':[], 'Amount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
              # print(D['data']['usersByDevice'])
                if D['data']['districts'] is not None:
                    for z in D['data']['districts']:
                        dist=z['entityName']
                        count=z['metric']['count']
                        amount=z['metric']['amount']
                        Top_transac['District'].append(dist)
                        Top_transac['Count'].append(count)
                        Top_transac['Amount'].append(amount)
                        Top_transac['State'].append(i)
                        Top_transac['Year'].append(j)
                        Top_transac['Quarter'].append(int(k.strip('.json')))
                  # print(brand)
      #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(Top_transac)

    if tableName == "aggTopTransaction":
      BulkloadDatatoTable("aggTopTransaction",Agg_Trans)
    elif tableName == "aggTopInsurance":
      BulkloadDatatoTable("aggTopInsurance",Agg_Trans)
  
  def Load_Top_user(self,path): # Instance Method
    Agg_state_list=os.listdir(path)
    Agg_state_list
    #Agg_state_list--> to get the list of states in India

    #<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

    #This is to extract the data's to create a dataframe

    TopUsers={'State':[], 'Year':[],'Quarter':[],'District':[],'Pincode':[], 'RegisteredCount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
              # print(D['data']['usersByDevice'])
                if D['data']['districts'] is not None:
                    for z in D['data']['districts']:
                        district=z['name']
                        registeredCount=z['registeredUsers']
                        TopUsers['District'].append(district)
                        TopUsers['Pincode'].append('')
                        TopUsers['RegisteredCount'].append(registeredCount)
                        TopUsers['State'].append(i)
                        TopUsers['Year'].append(j)
                        TopUsers['Quarter'].append(int(k.strip('.json')))
                    for z in D['data']['pincodes']:
                        pincode=z['name']
                        registeredCount=z['registeredUsers']
                        TopUsers['District'].append('')
                        TopUsers['Pincode'].append(pincode)
                        TopUsers['RegisteredCount'].append(registeredCount)
                        TopUsers['State'].append(i)
                        TopUsers['Year'].append(j)
                        TopUsers['Quarter'].append(int(k.strip('.json')))
                  # print(brand)
      #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(TopUsers)
    BulkloadDatatoTable("aggTopUsers",Agg_Trans)

  def Load_Map_TransactionInsurance(self,path,tableName): # Instance Method
    Agg_state_list=os.listdir(path)
    Agg_state_list
    #Agg_state_list--> to get the list of states in India

    #<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

    #This is to extract the data's to create a dataframe

    Map_transac={'State':[], 'Year':[],'Quarter':[],'District':[], 'Count':[], 'Amount':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
              # print(D['data']['usersByDevice'])
                if D['data']['hoverDataList'] is not None:
                    for z in D['data']['hoverDataList']:
                        dist=z['name']
                        count=z['metric'][0]['count']
                        amount=z['metric'][0]['amount']
                        Map_transac['District'].append(dist)
                        Map_transac['Count'].append(count)
                        Map_transac['Amount'].append(amount)
                        Map_transac['State'].append(i)
                        Map_transac['Year'].append(j)
                        Map_transac['Quarter'].append(int(k.strip('.json')))
                  # print(brand)
      #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(Map_transac)

    if tableName == "aggMapTransaction":
      BulkloadDatatoTable("aggMapTransaction",Agg_Trans)
    elif tableName == "aggMapInsurance":
      BulkloadDatatoTable("aggMapInsurance",Agg_Trans)
  
  def Load_Map_user(self,path): # Instance Method
    Agg_state_list=os.listdir(path)
    Agg_state_list
    #Agg_state_list--> to get the list of states in India

    #<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

    #This is to extract the data's to create a dataframe

    TopUsers={'State':[], 'Year':[],'Quarter':[],'District':[], 'RegisteredCount':[], 'AppOpens':[]}

    for i in Agg_state_list:
        p_i=path+i+"/"
        Agg_yr=os.listdir(p_i)
        for j in Agg_yr:
            p_j=p_i+j+"/"
            Agg_yr_list=os.listdir(p_j)
            for k in Agg_yr_list:
                p_k=p_j+k
                Data=open(p_k,'r')
                D=json.load(Data)
              # print(D['data']['usersByDevice'])
                if D['data']['hoverData'] is not None:
                    for dist, Values in D['data']['hoverData'].items():
                        district=dist
                        registeredCount=Values['registeredUsers']
                        appOpen=Values['appOpens']
                        TopUsers['District'].append(district)
                        TopUsers['RegisteredCount'].append(registeredCount)
                        TopUsers['AppOpens'].append(appOpen)
                        TopUsers['State'].append(i)
                        TopUsers['Year'].append(j)
                        TopUsers['Quarter'].append(int(k.strip('.json')))
                  # print(brand)
      #Succesfully created a dataframe
    Agg_Trans=pd.DataFrame(TopUsers)
    BulkloadDatatoTable("aggMapUsers",Agg_Trans)

def BulkloadDatatoTable(tableName,datafr):
   local_DB_URL="postgresql://postgres:admin@localhost:5432/Phonepe"
   try:
      datafr
      engine=create_engine(local_DB_URL)
      datafr.to_sql(tableName,engine,if_exists='replace',index=False,chunksize=500,method='multi')
      print('Data Inserted Successfully')
   except Exception as e:
      print(e)
```
